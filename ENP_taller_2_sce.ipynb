# 3

Verifique que el Kernel gaussiano cumple las propiedades de ser un Kernel

## Solución:

El kernel Gaussiano se define como:

$$
K(x) = \frac{1}{\sqrt{2\pi}} e^{-x^2/2}
$$

Verifiquemos cada propiedad de los kernel:

1. **$K(x) \geq 0$**: 

    Es facil demotrar que $K(x)$ siempre es $\geq 0$. Si $x \in \mathbb{R} $ entonces $x$ puede tomar valores negativos. Sin embargo, $x^2 \geq 0$ para cualquier valor de $x$. Por otro lado, $e^ {x^2}$  es una función eponencial, entonces $e^y \geq 0$ para cualquier valor de $y$. Es trivial mencionar que $1/\sqrt{2\pi} \geq 0$.
2. **$\int_{-\infty}^{\infty} K(x) \, dx = 1$**
    Primero, definamos la función error $\operatorname{erf}(x)$:

    $$
    \operatorname{erf}(x) = \frac{2}{\sqrt{\pi}} \int_{0}^{x} e^{-t^2} \, dt
    $$

    Ahora, consideremos la integral en todo el dominio del kernel gaussiano:

    $$
    \int_{-\infty}^{\infty} K(x) \, dx 
    $$ 

    $$
    =\frac{1}{\sqrt{2\pi}} \int_{-\infty}^{\infty}  e^{-x^2/2} \, dx 
    $$ 
    Considere un cambio de variable $u=x/\sqrt{2}$, luego, $du = dx/\sqrt{2}$.
    REmplazado:

    $$
    \frac{1}{\sqrt{2\pi}} \int_{-\infty}^{\infty}  e^{-x^2/2} \, dx = \frac{1}{\sqrt{2\pi}} \int_{-\infty}^{\infty}  e^{-u^2} \, (\sqrt{2}dx)
    $$ 
    $$
    \frac{\sqrt{2}}{\sqrt{2\pi}} \int_{-\infty}^{\infty}  e^{-u^2} \, dx 
    $$ 
    
    La función $e^{-u^2}$ es una función par, por lo tanto es simétrica, así:

    $$
    \frac{\sqrt{2}}{\sqrt{2\pi}} \int_{-\infty}^{\infty}  e^{-u^2} \, dx = \frac{\sqrt{2} }{\sqrt{2\pi}} (2 \int_{0}^{\infty}  e^{-u^2} \, dx)
    $$ 

    Ahora, $2 \int_{0}^{\infty}  e^{-u^2} \, dx = \sqrt{\pi}\operatorname{erf}(\infty)$. Luego
    $$
    \frac{\sqrt{2} }{\sqrt{2\pi}} (2 \int_{0}^{\infty}  e^{-u^2} \, dx) = \frac{\sqrt{2} \sqrt{\pi}}{\sqrt{2\pi}} \operatorname{erf}(\infty)
    $$ 

    pero $\operatorname{erf}(\infty) = 1$. Finalmente:

    $$
    \frac{\sqrt{2} \sqrt{\pi}}{\sqrt{2\pi}} = 1 = \int_{-\infty}^{\infty} K(x) \, dx
    $$ 
3. **$\int_{-\infty}^{\infty} xK(x) \, dx = 0$**

    $$
    \int_{-\infty}^{\infty} K(x) \, dx 
    $$ 

    $$
    =\frac{1}{\sqrt{2\pi}} \int_{-\infty}^{\infty}  xe^{-x^2/2} \, dx 
    $$ 

    Note que $xe^{-x^2/2}$ es una función impar, ya que $y(x)=x$ es una función impar, y el producto de una función par (en este caso $e^{-x^2/2}$) por función impar, es otra funión impar. Recuerde que $\int_{-\infty}^{\infty} f(x)  dx = 0$ para cualquier función $f(x)$ impar. Así
    $$
    \frac{1}{\sqrt{2\pi}} \int_{-\infty}^{\infty}  xe^{-x^2/2} \, dx = 0
    $$ 
4. **$\int_{-\infty}^{\infty} x^2K(x) \, > 0$**
    $$
    \int_{-\infty}^{\infty} K(x) \, dx 
    $$ 

    $$
    =\frac{1}{\sqrt{2\pi}} \int_{-\infty}^{\infty}  x^2e^{-x^2/2} \, dx 
    $$ 

    de forma similar al punto (2), se procede:

    $$
    \frac{1}{\sqrt{2\pi}} \int_{-\infty}^{\infty}  x^2e^{-x^2/2} \, dx 
    $$ 

    realizando el cambio de vriable $u = x/\sqrt{2}$, $du=dx/\sqrt{2}$, se tiene

    $$
    \frac{2\sqrt{2}}{\sqrt{2\pi}} \int_{-\infty}^{\infty}  u^2e^{-u^2} \, du 
    $$ 

    La integral $\int_{-\infty}^{\infty}  u^2e^{-u^2} \, dx $ se evalúa utilizando el hecho de que la función bajo la integral es par y aplicando un cambio de variable $v=u^2$, $dv = 2udu$:

    $$
    \int_{-\infty}^\infty u^2 e^{-u^2} \, du = 2 \int_0^\infty u^2 e^{-u^2} \, du= \int_0^\infty v^{1/2} e^{-v} \, dv
    $$
    La integral transformada representa la función gamma $\Gamma(\frac{3}{2})$:
    $$
    \int_0^\infty v^{1/2} e^{-v} \, dv = \Gamma\left(\frac{3}{2}\right)
    $$

    Utilizando las propiedades de la función gamma, específicamente la relación recursiva y conocimientos anteriores:
    $$
    \Gamma\left(\frac{3}{2}\right) = \frac{1}{2} \Gamma\left(\frac{1}{2}\right) = \frac{\sqrt{\pi}}{2}
    $$

    Así, la integral inicial es

    $$
    \frac{2}{\sqrt{\pi}} \int_{-\infty}^{\infty}  u^2e^{-u^2} \, du = \frac{2}{\sqrt{\pi}}\frac{\sqrt{\pi}}{2} = 1 > 0
    $$ 

    Lo cual es lo que se quería demostrar




# 4

De las siguientes afirmaciones, diga si son verdaderas o falsas justificando:

1. La suma de Kernels es un Kernel
2. Cualquier Kernel puede ser utilizado para generar una regresión no paramétrica

## Solución

### 1) La suma de Kernels es un Kernel

Consideremos dos kernels $K_1$ y $K_2$ que cumplen con las condiciones típicas de los kernels en estadística. La suma de estos kernels se define como $K(x) = K_1(x) + K_2(x)$. Analicemos si esta suma cumple con las condiciones de un kernel:

### 1. No negatividad: $K(x) \geq 0$

Ya que tanto $K_1(x) \geq 0$ como $K_2(x) \geq 0$ para todo $x$, la suma $K(x) = K_1(x) + K_2(x)$ también será no negativa en todos los puntos.

### 2. Integral igual a 1: $\int_{-\infty}^{\infty} K(x) \, dx = 1$

Al calcular la integral de la suma, tenemos:
$$\int_{-\infty}^{\infty} K(x) \, dx = \int_{-\infty}^{\infty} K_1(x) \, dx + \int_{-\infty}^{\infty} K_2(x) \, dx = 1 + 1 = 2.$$
Esto muestra que $K(x)$ no cumple esta condición, pues la integral debería sumar 1.

### 3. Media cero: $\int_{-\infty}^{\infty} xK(x) \, dx = 0$

Cada kernel tiene una media de 0, por lo tanto:
$$\int_{-\infty}^{\infty} xK(x) \, dx = \int_{-\infty}^{\infty} xK_1(x) \, dx + \int_{-\infty}^{\infty} xK_2(x) \, dx = 0 + 0 = 0.$$
La suma $K(x)$ sí cumple con esta condición.

### 4. Varianza positiva: $\int_{-\infty}^{\infty} x^2K(x) \, dx > 0$

Suponiendo independencia o al menos no correlación perfecta, las varianzas se suman:
$$\int_{-\infty}^{\infty} x^2K(x) \, dx = \int_{-\infty}^{\infty} x^2K_1(x) \, dx + \int_{-\infty}^{\infty} x^2K_2(x) \, dx > 0.$$
La suma también tiene una varianza positiva, cumpliendo esta condición.

**Conclusión**: La suma $K(x) = K_1(x) + K_2(x)$ cumple con todas las condiciones excepto la de tener una integral total igual a 1. Para ajustar esto y mantener la suma como un kernel válido, sería necesario escalar la suma por un factor de $\frac{1}{2}$, es decir, usar $K(x) = \frac{1}{2}(K_1(x) + K_2(x))$ para que la integral de la suma sea 1. Se concluye que la suma de kernels no es un kernel, pero la combinación lineal de kernels si puede ser un kernel, definiendo adecaudamente los coeficientes de cada kernel.

### 2) Cualquier Kernel puede ser utilizado para generar una regresión no paramétrica

Una de las tecnicas de regresión que implementa Kernels es el método de Nadaraya-Watson.
El método de regresión de Nadaraya-Watson es una técnica de regresión no paramétrica que utiliza funciones kernel para estimar la función de regresión.

La estimación de Nadaraya-Watson se calcula utilizando la siguiente fórmula:

$$
\hat{y}(x) = \frac{\sum_{i=1}^n K_h(x - x_i) y_i}{\sum_{i=1}^n K_h(x - x_i)}
$$
donde:
- $\hat{y}(x)$ es el valor estimado de $ y $ en $ x $
- $K_h$ es la función kernel, que depende de un parámetro de ancho de banda $h$. Este se define como $
K_h(x) = \frac{1}{h} K( \frac{x}{h})$
- $x_i$ y $y_i$ son los datos de entrenamiento.

Note que en esta definicón, $K(x)$ puede ser cualquier Kernel, tal como el gaussiano, uniforme, Boxcar, Epanechnikov, tricubico, triangular, entre otros.

Considere la siguiente función de densidad conjunta

$$
f(x, y) = 
\begin{cases} 
\frac{3}{16} (x^2 + y^2) & \text{si } 0 < x < y < 2 \\
0 & \text{en otro caso}
\end{cases}
$$

Realice la estimación de la regresión no paramétrica $E(Y|X)$. Adicionalmente, con el modelo de regresión teórico encontrado, realice la simulación de 1000 valores de la $Y$ agregando ruido blanco con desviación de 0.1. Posteriormente, realice la estimación de 1 modelo de regresión robusto y/o no paramétrico, además de la regresión lineal tradicional y compare el desempeño de haber estimado un modelo teórico con una versión no paramétrica VS una paramétrica.


**Solución:**

La función de densidad conjunta dada es:

$$
f(x, y) = \begin{cases} 
\frac{3}{16} (x^2 + y^2) & \text{si } 0 < x < y < 2 \\
0 & \text{en otro caso}
\end{cases}
$$

Para encontrar la función de densidad marginal de $X$, $f_X(x)$, integramos la función de densidad conjunta sobre todos los valores posibles de $y$, manteniendo $x$ fijo:

$$
f_X(x) = \int_{y=x}^{2} \frac{3}{16} (x^2 + y^2) \, dy
$$

Los límites de integración para $y$ son desde $x$ hasta $2$, porque $f(x, y)$ es diferente de cero solo cuando $0 < x < y < 2$.

Calculamos la integral de $x^2$ con respecto a $y$:

$$
\int x^2 \, dy = x^2 y \bigg|_{y=x}^{y=2} = x^2 (2 - x)
$$

Calculamos la integral de $y^2$ con respecto a $y$:

$$
\int y^2 \, dy = \frac{y^3}{3} \bigg|_{y=x}^{y=2} = \frac{2^3}{3} - \frac{x^3}{3} = \frac{8}{3} - \frac{x^3}{3}
$$

Combinando ambas partes:

$$
x^2(2 - x) + \left(\frac{8}{3} - \frac{x^3}{3}\right)
$$

Multiplicando por el factor constante $ \frac{3}{16} $ y simplificando:

$$
f_X(x) = \frac{3}{16} \left(2x^2 - x^3 + \frac{8}{3} - \frac{x^3}{3}\right)
= \frac{3x^2}{8} - \frac{x^3}{12} + \frac{1}{2}, \quad \text{si } 0 < x < 2
$$

En cualquier otro caso:

$$
f_X(x) = 0
$$

Por lo tanto, la función de densidad marginal de $ X $ es:

$$
f_X(x) = \begin{cases} 
\frac{3x^2}{8} - \frac{x^3}{12} + \frac{1}{2} & \text{si } 0 < x < 2 \\
0 & \text{en otro caso}
\end{cases}
$$


Para calcular la función de densidad de probabilidad de $ Y $ condicionada a $ X = x $, usamos la fórmula:

$$
f_{Y|X}(y|x) = \frac{f(x, y)}{f_X(x)}
$$

Esto es

$$
f_{Y|X}(y|x) = \frac{\frac{3}{16} (x^2 + y^2)}{\frac{3x^2}{8} - \frac{x^3}{12} + \frac{1}{2}}
$$

y simplificando

$$
f_{Y|X}(y|x) = \frac{9(x^2 + y^2)}{2(9x^2 - 2x^3 + 12)}
$$

Para calcular el valor esperado de $ Y $ dado $ X = x $, $E(Y | X = x) $ usamos la definición de valor esperado para una variable aleatoria continua. Este se calcula como la integral del producto de la variable y su función de densidad condicionada:

$$
E(Y | X = x) = \int_{-\infty}^\infty y \cdot f_{Y|X}(y|x) \, dy
$$

Dada la función de densidad de probabilidad de $ Y $ condicionada a $ X = x $:

$$
f_{Y|X}(y|x) = \frac{9(x^2 + y^2)}{2(9x^2 - 2x^3 + 12)}
$$

Ajustamos los límites de integración de acuerdo a la región donde $f_{Y|X}(y|x) $ es no nula, es decir, cuando $0 < x < y < 2 $:

$$
E(Y | X = x) = \int_{x}^{2} y \cdot \frac{9(x^2 + y^2)}{2(9x^2 - 2x^3 + 12)} \, dy
$$

Desglosamos y simplificamos la integral:

$$
\int_{x}^{2} y \cdot \frac{9(x^2 + y^2)}{2(9x^2 - 2x^3 + 12)} \, dy = \frac{9}{2(9x^2 - 2x^3 + 12)} \int_{x}^{2} y(x^2 + y^2) \, dy
$$

Simplificando más, tenemos:

$$
\int_{x}^{2} (yx^2 + y^3) \, dy = x^2 \int_{x}^{2} y \, dy + \int_{x}^{2} y^3 \, dy
$$

$$
= x^2 \left[\frac{y^2}{2}\right]_{x}^{2} + \left[\frac{y^4}{4}\right]_{x}^{2}
$$

$$
= x^2 \left(\frac{4}{2} - \frac{x^2}{2}\right) + \left(\frac{16}{4} - \frac{x^4}{4}\right)
$$

$$
= 2x^2 - \frac{x^4}{2} + 4 - \frac{x^4}{4}
$$

Insertando esto de vuelta en la fórmula del valor esperado:

$$
E(Y | X = x) = \frac{9}{2(9x^2 - 2x^3 + 12)} \left(2x^2 - 3\frac{x^4}{4} + 4\right)
$$

simplificando los más posible:

$$
E(Y|X = x) = \frac{9(x - 2)(x + 2)(3x^2 + 4)}{8x^2(2x - 9) - 96}
$$

Esta expresión representa el valor esperado de $ Y $ dado $ X = x $ dentro del intervalo permitido, que es crucial para entender cómo $Y $ se comporta en promedio dado un valor específico de $X $ en el contexto definido por las funciones de densidad proporcionadas.




```python
import numpy as np
import plotly.graph_objects as go

# Definir la función f(x, y)
def f(x, y):
    return (3/16) * (x**2 + y**2) if 0 < x < y < 2 else 0

# Crear un grid de valores x, y
x = np.linspace(0, 2, 100)
y = np.linspace(0, 2, 100)
X, Y = np.meshgrid(x, y)
Z = np.zeros_like(X)

# Calcular los valores Z basados en la condición de f(x, y)
for i in range(len(x)):
    for j in range(len(y)):
        if 0 < X[i, j] < Y[i, j] < 2:
            Z[i, j] = f(X[i, j], Y[i, j])

# Crear la figura usando Plotly
fig = go.Figure(data=[go.Surface(z=Z, x=X, y=Y)])

# Actualizar el layout de la figura
fig.update_layout(title='Visualización Interactiva de la Función f(x, y)',
                  scene=dict(
                      xaxis_title='X',
                      yaxis_title='Y',
                      zaxis_title='f(x, y)'),
                  autosize=False,
                  width=800, height=800,
                  margin=dict(l=65, r=50, b=65, t=90))

# Mostrar la figura
fig.show()
```

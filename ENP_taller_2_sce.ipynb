# 3

Verifique que el Kernel gaussiano cumple las propiedades de ser un Kernel

## Solución:

El kernel Gaussiano se define como:

$$
K(x) = \frac{1}{\sqrt{2\pi}} e^{-x^2/2}
$$

Verifiquemos cada propiedad de los kernel:

1. **$K(x) \geq 0$**: 

    Es facil demotrar que $K(x)$ siempre es $\geq 0$. Si $x \in \mathbb{R} $ entonces $x$ puede tomar valores negativos. Sin embargo, $x^2 \geq 0$ para cualquier valor de $x$. Por otro lado, $e^ {x^2}$  es una función eponencial, entonces $e^y \geq 0$ para cualquier valor de $y$. Es trivial mencionar que $1/\sqrt{2\pi} \geq 0$.
2. **$\int_{-\infty}^{\infty} K(x) \, dx = 1$**
    Primero, definamos la función error $\operatorname{erf}(x)$:

    $$
    \operatorname{erf}(x) = \frac{2}{\sqrt{\pi}} \int_{0}^{x} e^{-t^2} \, dt
    $$

    Ahora, consideremos la integral en todo el dominio del kernel gaussiano:

    $$
    \int_{-\infty}^{\infty} K(x) \, dx 
    $$ 

    $$
    =\frac{1}{\sqrt{2\pi}} \int_{-\infty}^{\infty}  e^{-x^2/2} \, dx 
    $$ 
    Considere un cambio de variable $u=x/\sqrt{2}$, luego, $du = dx/\sqrt{2}$.
    REmplazado:

    $$
    \frac{1}{\sqrt{2\pi}} \int_{-\infty}^{\infty}  e^{-x^2/2} \, dx = \frac{1}{\sqrt{2\pi}} \int_{-\infty}^{\infty}  e^{-u^2} \, (\sqrt{2}dx)
    $$ 
    $$
    \frac{\sqrt{2}}{\sqrt{2\pi}} \int_{-\infty}^{\infty}  e^{-u^2} \, dx 
    $$ 
    
    La función $e^{-u^2}$ es una función par, por lo tanto es simétrica, así:

    $$
    \frac{\sqrt{2}}{\sqrt{2\pi}} \int_{-\infty}^{\infty}  e^{-u^2} \, dx = \frac{\sqrt{2} }{\sqrt{2\pi}} (2 \int_{0}^{\infty}  e^{-u^2} \, dx)
    $$ 

    Ahora, $2 \int_{0}^{\infty}  e^{-u^2} \, dx = \sqrt{\pi}\operatorname{erf}(\infty)$. Luego
    $$
    \frac{\sqrt{2} }{\sqrt{2\pi}} (2 \int_{0}^{\infty}  e^{-u^2} \, dx) = \frac{\sqrt{2} \sqrt{\pi}}{\sqrt{2\pi}} \operatorname{erf}(\infty)
    $$ 

    pero $\operatorname{erf}(\infty) = 1$. Finalmente:

    $$
    \frac{\sqrt{2} \sqrt{\pi}}{\sqrt{2\pi}} = 1 = \int_{-\infty}^{\infty} K(x) \, dx
    $$ 
3. **$\int_{-\infty}^{\infty} xK(x) \, dx = 0$**

    $$
    \int_{-\infty}^{\infty} K(x) \, dx 
    $$ 

    $$
    =\frac{1}{\sqrt{2\pi}} \int_{-\infty}^{\infty}  xe^{-x^2/2} \, dx 
    $$ 

    Note que $xe^{-x^2/2}$ es una función impar, ya que $y(x)=x$ es una función impar, y el producto de una función par (en este caso $e^{-x^2/2}$) por función impar, es otra funión impar. Recuerde que $\int_{-\infty}^{\infty} f(x)  dx = 0$ para cualquier función $f(x)$ impar. Así
    $$
    \frac{1}{\sqrt{2\pi}} \int_{-\infty}^{\infty}  xe^{-x^2/2} \, dx = 0
    $$ 
4. **$\int_{-\infty}^{\infty} x^2K(x) \, > 0$**
    $$
    \int_{-\infty}^{\infty} K(x) \, dx 
    $$ 

    $$
    =\frac{1}{\sqrt{2\pi}} \int_{-\infty}^{\infty}  x^2e^{-x^2/2} \, dx 
    $$ 

    de forma similar al punto (2), se procede:

    $$
    \frac{1}{\sqrt{2\pi}} \int_{-\infty}^{\infty}  x^2e^{-x^2/2} \, dx 
    $$ 

    realizando el cambio de vriable $u = x/\sqrt{2}$, $du=dx/\sqrt{2}$, se tiene

    $$
    \frac{2\sqrt{2}}{\sqrt{2\pi}} \int_{-\infty}^{\infty}  u^2e^{-u^2} \, du 
    $$ 

    La integral $\int_{-\infty}^{\infty}  u^2e^{-u^2} \, dx $ se evalúa utilizando el hecho de que la función bajo la integral es par y aplicando un cambio de variable $v=u^2$, $dv = 2udu$:

    $$
    \int_{-\infty}^\infty u^2 e^{-u^2} \, du = 2 \int_0^\infty u^2 e^{-u^2} \, du= \int_0^\infty v^{1/2} e^{-v} \, dv
    $$
    La integral transformada representa la función gamma $\Gamma(\frac{3}{2})$:
    $$
    \int_0^\infty v^{1/2} e^{-v} \, dv = \Gamma\left(\frac{3}{2}\right)
    $$

    Utilizando las propiedades de la función gamma, específicamente la relación recursiva y conocimientos anteriores:
    $$
    \Gamma\left(\frac{3}{2}\right) = \frac{1}{2} \Gamma\left(\frac{1}{2}\right) = \frac{\sqrt{\pi}}{2}
    $$

    Así, la integral inicial es

    $$
    \frac{2}{\sqrt{\pi}} \int_{-\infty}^{\infty}  u^2e^{-u^2} \, du = \frac{2}{\sqrt{\pi}}\frac{\sqrt{\pi}}{2} = 1 > 0
    $$ 

    Lo cual es lo que se quería demostrar




# 4

De las siguientes afirmaciones, diga si son verdaderas o falsas justificando:

1. La suma de Kernels es un Kernel
2. Cualquier Kernel puede ser utilizado para generar una regresión no paramétrica

## Solución

### 1) La suma de Kernels es un Kernel

Consideremos dos kernels $K_1$ y $K_2$ que cumplen con las condiciones típicas de los kernels en estadística. La suma de estos kernels se define como $K(x) = K_1(x) + K_2(x)$. Analicemos si esta suma cumple con las condiciones de un kernel:

### 1. No negatividad: $K(x) \geq 0$

Ya que tanto $K_1(x) \geq 0$ como $K_2(x) \geq 0$ para todo $x$, la suma $K(x) = K_1(x) + K_2(x)$ también será no negativa en todos los puntos.

### 2. Integral igual a 1: $\int_{-\infty}^{\infty} K(x) \, dx = 1$

Al calcular la integral de la suma, tenemos:
$$\int_{-\infty}^{\infty} K(x) \, dx = \int_{-\infty}^{\infty} K_1(x) \, dx + \int_{-\infty}^{\infty} K_2(x) \, dx = 1 + 1 = 2.$$
Esto muestra que $K(x)$ no cumple esta condición, pues la integral debería sumar 1.

### 3. Media cero: $\int_{-\infty}^{\infty} xK(x) \, dx = 0$

Cada kernel tiene una media de 0, por lo tanto:
$$\int_{-\infty}^{\infty} xK(x) \, dx = \int_{-\infty}^{\infty} xK_1(x) \, dx + \int_{-\infty}^{\infty} xK_2(x) \, dx = 0 + 0 = 0.$$
La suma $K(x)$ sí cumple con esta condición.

### 4. Varianza positiva: $\int_{-\infty}^{\infty} x^2K(x) \, dx > 0$

Suponiendo independencia o al menos no correlación perfecta, las varianzas se suman:
$$\int_{-\infty}^{\infty} x^2K(x) \, dx = \int_{-\infty}^{\infty} x^2K_1(x) \, dx + \int_{-\infty}^{\infty} x^2K_2(x) \, dx > 0.$$
La suma también tiene una varianza positiva, cumpliendo esta condición.

**Conclusión**: La suma $K(x) = K_1(x) + K_2(x)$ cumple con todas las condiciones excepto la de tener una integral total igual a 1. Para ajustar esto y mantener la suma como un kernel válido, sería necesario escalar la suma por un factor de $\frac{1}{2}$, es decir, usar $K(x) = \frac{1}{2}(K_1(x) + K_2(x))$ para que la integral de la suma sea 1. Se concluye que la suma de kernels no es un kernel, pero la combinación lineal de kernses si puede ser un kernel, definiendo adecaudamente los coeficientes de cada kernel.

# Evaluación de la Adecuación de Cualquier Kernel para Regresión No Paramétrica

## Propiedades Fundamentales de los Kernels

Un kernel en el contexto de la regresión no paramétrica debe satisfacer varias propiedades clave para ser efectivo:

1. **No negatividad**: $K(x) \geq 0$ para todos los valores de $x$. Esto asegura que los pesos asignados a los puntos de datos no sean negativos, lo cual tiene sentido desde una perspectiva de ponderación de datos.

2. **Integración a uno**: $\int_{-\infty}^{\infty} K(x) \, dx = 1$. Esta propiedad garantiza que el kernel actúe como una función de densidad de probabilidad, asignando efectivamente una distribución de probabilidad sobre los datos.

3. **Simetría**: Generalmente, los kernels son simétricos alrededor del punto central, es decir, $K(x) = K(-x)$. Esto asegura que la influencia de los datos sobre la estimación sea equitativa en ambas direcciones desde cualquier punto dado.

## Aplicabilidad de Cualquier Kernel en Regresión No Paramétrica

### Teoría

En teoría, cualquier función que cumpla con estas propiedades puede ser considerada un kernel y ser utilizada en regresión no paramétrica. La elección del kernel afecta la suavidad y la forma del ajuste, pero mientras mantenga las propiedades básicas, puede ser utilizado para ponderar los datos en la regresión.

### Práctica

En la práctica, sin embargo, no todos los kernels pueden ser igualmente efectivos para todos los tipos de datos o problemas. La elección del kernel y su ancho de banda asociado (la ventana de efecto) son críticos y pueden influir significativamente en la calidad del modelo de regresión no paramétrica:

- **Kernels estrechos** pueden llevar a un modelo que se ajusta demasiado a los datos (overfitting), capturando el "ruido" como si fuera parte de la estructura subyacente del fenómeno.
- **Kernels amplios** pueden suavizar demasiado los datos, perdiendo características importantes y llevando a un modelo subajustado (underfitting).

### Conclusión

Mientras que teóricamente cualquier kernel que cumpla con las propiedades básicas puede ser utilizado para la regresión no paramétrica, la efectividad del análisis depende fuertemente de la elección apropiada del tipo de kernel y su ancho de banda en relación con los datos específicos y el contexto del problema. Por lo tanto, es crucial realizar una selección informada y, a menudo, ajustes basados en la validación cruzada u otros métodos de evaluación de modelos para determinar la configuración óptima.
